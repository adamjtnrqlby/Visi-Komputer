```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('aksarajawa.png', 0)  # 0 = grayscale
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')
```

```python
kernel = np.ones((5,5), np.uint8)  # ukuran kernel 5x5
erosi = cv2.erode(img, kernel, iterations=1)
```

```python
plt.imshow(erosi, cmap='gray')
plt.title('Hasil Erosi')
plt.axis('off')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(erosi, cmap='gray')
plt.title('Erosi')
plt.axis('off')
```
