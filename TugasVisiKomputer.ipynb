# **a. Erosi**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
img = cv2.imread('aksarajawa.png', 0)  # 0 = grayscale
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')
```

```python
kernel = np.ones((5,5), np.uint8)  # ukuran kernel 5x5
erosi = cv2.erode(img, kernel, iterations=1)
```

```python
plt.imshow(erosi, cmap='gray')
plt.title('Hasil Erosi')
plt.axis('off')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(erosi, cmap='gray')
plt.title('Erosi')
plt.axis('off')
```

# **b. Opening**

```python
img = cv2.imread('aksarajawa.png', 0)
```

```python
kernel = np.ones((5,5), np.uint8)
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening, cmap='gray')
plt.title('Hasil Opening')
plt.axis('off')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.axis('off')
```

```python
plt.figure(figsize=(15,5))

# Gambar Asli
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Asli')
plt.axis('off')

# Hasil Erosi
plt.subplot(1,3,2)
plt.imshow(erosi, cmap='gray')
plt.title('Erosi')
plt.axis('off')

# Hasil Opening
plt.subplot(1,3,3)
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.axis('off')

plt.show()
```

# c. Closing

```python
img = cv2.imread('aksarajawa.png', 0)
kernel = np.ones((5,5), np.uint8)
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closing, cmap='gray')
plt.title('Hasil Closing')
plt.axis('off')
```

```python
plt.figure(figsize=(15,8))

# Asli
plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Asli')
plt.axis('off')

# Erosi
plt.subplot(2,2,2)
plt.imshow(erosi, cmap='gray')
plt.title('Erosi')
plt.axis('off')

# Opening
plt.subplot(2,2,3)
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.axis('off')

# Closing
plt.subplot(2,2,4)
plt.imshow(closing, cmap='gray')
plt.title('Closing')
plt.axis('off')

plt.show()
```
